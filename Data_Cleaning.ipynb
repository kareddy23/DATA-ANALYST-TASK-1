```python
import pandas as pd
```

```python
# Load the dataset
file_path = "/content/marketing_campaign.csv"
df = pd.read_csv(file_path)
```

```python
# Reload the dataset with tab separator
df = pd.read_csv(file_path, sep="\t")

# Display the first few rows again to confirm proper loading
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Fill missing Income values with the median
df["Income"].fillna(df["Income"].median(), inplace=True)

# Remove duplicate rows, if any
df_before = df.shape[0]
df.drop_duplicates(inplace=True)
df_after = df.shape[0]
duplicates_removed = df_before - df_after

duplicates_removed
```
